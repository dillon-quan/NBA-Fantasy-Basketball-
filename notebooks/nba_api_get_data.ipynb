### Player Data

```python
# players descriptive data
df_players = pd.DataFrame(players.get_players())
df_players.rename(columns={'id':'player_id'}, inplace=True)
df_players.head()
```

### Career Stats

* 0 - overall regular stats
* 2 - overall playoff stats
* 4 - overall allstar stats
* 6 - overall college stats

```python
curry_id = df_players.loc[(df_players.full_name.str.contains('Stephen Curry', regex=True, case=False))]['player_id'].values[0]
```

```python
la_id = df_players.loc[(df_players.full_name.str.contains('LaMarcus Aldridge', regex=True, case=False))]['player_id'].values[0]
la_id
```

```python
df_players.loc[(df_players.full_name.str.contains('Dario', regex=True, case=False))]
```

```python
# players career stats
from nba_api.stats.endpoints import playercareerstats
curry_career = playercareerstats.PlayerCareerStats(
    player_id=curry_id,
    per_mode36='PerGame'
)
curry_career.get_data_frames()[0]
```

```python
curry_career.get_data_frames()[2]
```

```python
curry_career.get_data_frames()[6]
```
