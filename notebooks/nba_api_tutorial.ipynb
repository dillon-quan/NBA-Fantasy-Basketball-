### Getting Player & Team Data

```python
from nba_api.stats.static import players
```

```python
# returns a list of dictionary of all players
players_dict = players.get_players()
len(players_dict)
```

```python
from nba_api.stats.static import teams
teams_dict = teams.get_teams()
len(teams_dict)
```

```python
teams_dict
```

### Finding Games data

```python
gsw_id = [team['id'] for team in teams_dict if team['abbreviation'] == 'GSW']
gsw_id[0]
```

```python
from nba_api.stats.endpoints import leaguegamefinder

game_finder = leaguegamefinder.LeagueGameFinder(team_id_nullable=gsw_id[0])
games = game_finder.get_data_frames()[0]
```

```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_columns = 30
```
